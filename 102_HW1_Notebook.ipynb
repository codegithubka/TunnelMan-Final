```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time values from -10 to 10 with a step size of 0.2
t = np.arange(-10, 10.2, 0.2)

# Calculate the corresponding function values
y = np.exp(-t) * np.cos(2 * np.pi * t)

# Create a labeled and titled plot
plt.figure(figsize=(8, 6))
plt.plot(t, y, label='$e^{-t} \cos(2\pi t)$')
plt.xlabel('Time $t$')
plt.ylabel('Amplitude')
plt.title('Plot of $e^{-t} \cos(2\pi t)$')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()






```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(t):
    return np.maximum(0, t)

# Define the time values from -5 to 5 with a step size of 0.1
t = np.arange(-5, 5.1, 0.1)

# Calculate the corresponding function values using the relu function
y = relu(t)

# Create a labeled and titled plot
plt.figure(figsize=(8, 6))
plt.plot(t, y, label='ReLU($t$)')
plt.xlabel('Time $t$')
plt.ylabel('Amplitude')
plt.title('Plot of ReLU($t$)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()




```

```python
# Define the ReLU function
def relu(t):
    return np.maximum(0, t)

# Define the even part of a function f(t)
def even(t, f):
    return 0.5 * (f(t) + f(-t))

# Define the odd part of a function f(t)
def odd(t, f):
    return 0.5 * (f(t) - f(-t))

# Define the time values from -5 to 5 with a step size of 0.1
t = np.arange(-5, 5.1, 0.1)

# Calculate the even and odd parts of the ReLU function
even_part = even(t, relu)
odd_part = odd(t, relu)

# Create labeled and titled plots for the even and odd parts
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, even_part, label='Even Part')
plt.xlabel('Time $t$')
plt.ylabel('Amplitude')
plt.title('Even Part of ReLU($t$)')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(t, odd_part, label='Odd Part', color='orange')
plt.xlabel('Time $t$')
plt.ylabel('Amplitude')
plt.title('Odd Part of ReLU($t$)')
plt.grid(True)
plt.legend()

plt.tight_layout()

# Show the plots
plt.show()

```
